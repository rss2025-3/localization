```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
sensor_model_data = np.load('sensor_model_table.npy')
```

```python
# Create meshgrid for the plot
x = np.arange(sensor_model_data.shape[0])  # Measured Distance
y = np.arange(sensor_model_data.shape[1])  # Ground Truth Distance
X, Y = np.meshgrid(x, y)

# Create the 3D plot
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, sensor_model_data.T, 
                      cmap='viridis',
                      linewidth=0,
                      antialiased=True)

# Set labels
ax.set_xlabel('Measured Distance (in px)')
ax.set_ylabel('Ground truth distance (in px)')
ax.set_zlabel('P(Measured Distance | Ground Truth)')

# Set title
plt.title('Precomputed Sensor Model')

# Add a color bar
fig.colorbar(surf)

# Adjust the viewing angle to match the example
ax.view_init(elev=20, azim=120)  # Changed azim from -45 to 45 to rotate 90 degrees

plt.savefig("senor_model.jpg", dpi=300)
#plt.show()
```
